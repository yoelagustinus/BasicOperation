# Import Modules

```python
import cv2
import matplotlib.pyplot as plt
```

***
# Display all original image

### - Lenna Image

```python
img_lenna = cv2.cvtColor(cv2.imread('../images/lenna.png'), cv2.COLOR_BGR2RGB)

plt.imshow(img_lenna)
plt.title('Lenna')
plt.show()

print(f'Height        : {img_lenna.shape[0]} pixels')
print(f'Width         : {img_lenna.shape[1]} pixels')
print(f'Color Channel : {img_lenna.shape[2]} (RGB))')
```
